```python
## import PuLP
from pulp import *

# Create the 'prob' variable to contain the problem data
prob = LpProblem("Let's cut some rolls", LpMinimize)

# Create problem variables
## Goal = 97 rolls of 135cm, 610rolls of 108cm, 395 rolls of 93cm, 211 rolls of 42cm
##One 135cm roll
x1=LpVariable("cut_typ_1",0,None,LpInteger)
##two 135cm rolls
x2=LpVariable("cut_typ_2",0, None, LpInteger)
##One 108 cm
x3=LpVariable("cut_typ_3",0, None, LpInteger)
##two 108 cm
x4=LpVariable("cut_typ_4",0, None, LpInteger)
##One 93cm
x5=LpVariable("cut_typ_5",0, None, LpInteger)
##two 93cm
x6=LpVariable("cut_typ_6",0,None,LpInteger)
##three 93cm
x7=LpVariable("cut_typ_7",0, None, LpInteger)
##one 42cm
x8=LpVariable("cut_typ_8",0, None, LpInteger)
##two 42cm
x9=LpVariable("cut_typ_9",0, None, LpInteger)
##three 42cm
x10=LpVariable("cut_typ_10",0, None, LpInteger)
##four 42cm
x11=LpVariable("cut_typ_11",0,None,LpInteger)
##five 42cm
x12=LpVariable("cut_typ_12",0, None, LpInteger)
##six 42cm 
x13=LpVariable("cut_typ_13",0, None, LpInteger)
##seven 42cm
x14=LpVariable("cut_typ_14",0, None, LpInteger)
##One 135cm rolls + one 108cm roll
x15=LpVariable("cut_typ_15",0, None, LpInteger)
##One 135cm rolls + one 108cm roll + one 42 cm
x16=LpVariable("cut_typ_16",0,None,LpInteger)
##One 135cm rolls + one 93cm roll 
x17=LpVariable("cut_typ_17",0, None, LpInteger)
##One 135cm rolls + One 93cm + one 42cm
x18=LpVariable("cut_typ_18",0, None, LpInteger)
##One 135cm rolls + One 42cm
x19=LpVariable("cut_typ_19",0, None, LpInteger)
##One 135cm rolls + Two 42cm
x20=LpVariable("cut_typ_20",0, None, LpInteger)
##One 135cm rolls + Three 42cm
x21=LpVariable("cut_typ_21",0,None,LpInteger)
##One 108cm roll + one 93cm roll
x22=LpVariable("cut_typ_22",0, None, LpInteger)
##One 108cm roll + two 93cm roll
x23=LpVariable("cut_typ_23",0, None, LpInteger)
##One 108cm roll + One 93cm roll + 42cm
x24=LpVariable("cut_typ_24",0, None, LpInteger)
##One 108cm roll + One 93cm roll + two 42m roll 
x25=LpVariable("cut_typ_25",0, None, LpInteger)
##One 108cm roll + One 42cm roll
x26=LpVariable("cut_typ_26",0,None,LpInteger)
##One 108cm roll + Two 42cm roll
x27=LpVariable("cut_typ_27",0, None, LpInteger)
##One 108cm roll + Thee 42cm roll
x28=LpVariable("cut_typ_28",0, None, LpInteger)
##One 108cm roll + Four 42cm roll
x29=LpVariable("cut_typ_29",0, None, LpInteger)
##Two 108cm roll + One 42cm roll
x30=LpVariable("cut_typ_30",0, None, LpInteger)
##TWo 108cm roll + TWo 42cm roll
x31=LpVariable("cut_typ_31",0, None, LpInteger)
##One 93cm roll + 42cm roll
x32=LpVariable("cut_typ_32",0, None, LpInteger)
##One 93cm roll + Two 42cm roll
x33=LpVariable("cut_typ_33",0,None,LpInteger)
##One 93cm roll + Three 42cm roll
x34=LpVariable("cut_typ_34",0, None, LpInteger)
##One 93cm roll + Four 42cm roll
x35=LpVariable("cut_typ_35",0, None, LpInteger)
##Two 93cm roll + One 42cm roll
x36=LpVariable("cut_typ_36",0, None, LpInteger)
##Two 93cm roll + Two 42cm roll
x37=LpVariable("cut_typ_37",0, None, LpInteger)


# The objective function is added to 'prob' first
prob += x1+x2+x3+x4+x5+x6+x7+x8+x9+x10+x11+x12+x13+x14+x15+x16+x17+x18+x19+x20+x21+x22+x23+x24+x25+x26+x27+x28+x29+x30+x31+x32+x33+x34+x35+x36+x37, "min"
# The two constraints are entered
prob += x1 + 2*x2+ x15+x16+x17+x18+x19+x20+x21 ==97, "Roll 1 Demand"
prob += x3+  2*x4+ x14+x15+x16+x22+x23+x24+x25+x26+x27+x28+x29 +2*x30+ 2*x31 ==610, "Roll 2 Demand"
prob += x5+2*x6+3*x7+ x17+x18+x22+2*x23+x24+x25+x32+x33+x34+x35+2*x36+2*x37 ==305, "Roll 3 Demand"
prob += x8+2*x9+3*x10+4*x11+5*x12+6*x13+7*x14 + x16+x18+x19+2*x20+3*x21+x24+2*x25+x26+2*x27+3*x28+4*x29+x30+2*x31+x32+2*x33+3*x34+4*x35+x36+2*x37 ==211, "Roll 4 Demand"
        
# The problem data is written to an .lp file
prob.writeLP("Cut_Paper.lp")

# The problem is solved using PuLP's choice of Solver
prob.solve()
```

```python
# The status of the solution is printed to the screen
print("Status:", LpStatus[prob.status])
# Output= 
# Status: Optimal

# Each of the variables is printed with it's resolved optimum value
for v in prob.variables():
    print(v.name, "=", v.varValue)
# Output=
# Medicine_1_units = 3.0
# Medicine_2_units = 4.0

# The optimised objective function value is printed to the screen
print("Total Health that can be restored = ", value(prob.objective))
# Output= 
# Total Health that can be restored =  155.0
```
